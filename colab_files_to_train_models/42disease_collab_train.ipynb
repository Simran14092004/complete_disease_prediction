```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import numpy as np
```

```python
# Load the training and testing datasets
train_df = pd.read_csv("/content/genericTraining.csv")
test_df = pd.read_csv("/content/genericTesting.csv")
```

```python
# Clean up any unnamed or empty columns
train_df = train_df.loc[:, ~train_df.columns.str.contains('^Unnamed')]
test_df = test_df.loc[:, ~test_df.columns.str.contains('^Unnamed')]
```

```python
# Preview
print("Training Data Shape:", train_df.shape)
```

```python
print("Testing Data Shape:", test_df.shape)
```

```python
# Show the first few rows
print("Dataset Preview:")
print(train_df.head())
```

```python
# Check for missing values
print("\nMissing values in training data:")
print(train_df.isnull().sum().sum() if train_df.isnull().values.any() else "No missing values in training set!")
```

```python
print("\nMissing values in testing data:")
print(test_df.isnull().sum().sum() if test_df.isnull().values.any() else "No missing values in testing set!")
```

```python
# Features and target
X_train = train_df.drop('prognosis', axis=1)
y_train = train_df['prognosis']
```

```python
X_test = test_df.drop('prognosis', axis=1)
y_test = test_df['prognosis']
```

```python
# Encode the target labels
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)
```

```python
# Train a Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train_encoded)
```

```python
# Train the model
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train_encoded)
```

```python
# Make predictions
y_pred = clf.predict(X_test)
```

```python
# Evaluate
print("\nAccuracy:", accuracy_score(y_test_encoded, y_pred))
```

```python
print("\nClassification Report:")
print(classification_report(y_test_encoded, y_pred, target_names=le.classes_))
```

```python
# Predict on training data
y_train_pred = clf.predict(X_train)
```

```python
train_accuracy = accuracy_score(y_train_encoded, y_train_pred)

print("Training Accuracy:", train_accuracy)
```

```python
print("Classification Report (Training Data):\n")
print(classification_report(y_train_encoded, y_train_pred, target_names=le.classes_))
```
