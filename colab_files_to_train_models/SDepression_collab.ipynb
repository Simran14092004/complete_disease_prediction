Importing the dependencies:


```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Get a list of numerical columns
numerical_cols = Sdepression_datadet.select_dtypes(include=np.number).columns

# Calculate the mean for numerical columns only
Sdepression_datadet.groupby('Depression')[numerical_cols].mean()
```

```python
# separating the data and labels
X = Sdepression_datadet.drop(columns = 'Depression', axis=1)
Y = Sdepression_datadet['Depression']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Step 2: Identify categorical columns
categorical_cols = X.select_dtypes(include='object').columns
```

```python
# Step 3: Convert categorical columns using one-hot encoding (or use LabelEncoding for binary)
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)
```

Data Standardization:

```python
# Impute missing values (only for numerical columns)

imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X_encoded)  # Impute missing values in the entire datase
```

```python
# 4. Scale the imputed data (Standardize the features)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)  # Scale after imputation
```

```python
# 5. Convert back to DataFrame for readability
X_scaled_df = pd.DataFrame(X_scaled, columns=X_encoded.columns)
```

```python
print(X_scaled_df.head())  # Check the first few rows of the scaled data
```

```python
print(Y.head())
```

Splitting the data to training data and test data:


```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled_df, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
# checking if the split for test and train:
print("Train label distribution:\n", Y_train.value_counts(normalize=True))
```

```python
print("Test label distribution:\n", Y_test.value_counts(normalize=True))
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Data Standardization:

```python
print(X_train)
```

Model training:

Support vector Machine:

```python
model = svm.SVC(kernel='linear')
```

```python
print(X_train.isna().sum().sort_values(ascending=False).head(10))  # to check if have any missing values(nan)
```

```python
#Training the svm model with training data
model.fit(X_train, Y_train)
```

Model Evaluation:
Acurracy Score:

```python
# accuracy Score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy score of training data:", training_data_accuracy )
```

```python
X_test_pred = model.predict(X_test)
test_acc = accuracy_score(Y_test, X_test_pred)
print("Accuracy score on test data:", test_acc)
```

Building a predictive system:

```python
# Your raw input data (make sure it's in the same order as the original DataFrame)
input_data = (62,'Male',31.0,'Nashik','Student',2.0,0.0,8.38,3.0,0.0,'Less than 5 hours','Moderate','LLB','Yes',2.0,5.0,'No')

# Convert to DataFrame (column names must match original dataset!)
input_df = pd.DataFrame([input_data], columns=[
    'id', 'Gender', 'Age', 'City', 'Profession', 'Academic Pressure', 'Work Pressure',
    'CGPA', 'Study Satisfaction', 'Job Satisfaction', 'Sleep Duration', 'Dietary Habits',
    'Degree', 'Have you ever had suicidal thoughts ?', 'Work/Study Hours',
    'Financial Stress', 'Family History of Mental Illness'
])

# Encode the categorical variables (use same one-hot encoding logic)
input_encoded = pd.get_dummies(input_df)

# Match the input to the training data's columns (fill missing dummy columns with 0)
input_encoded = input_encoded.reindex(columns=X_encoded.columns, fill_value=0)

# Impute missing values (optional if none are expected)
input_imputed = imputer.transform(input_encoded)

# Scale using the previously fitted scaler
input_scaled = scaler.transform(input_imputed)

# Make prediction
prediction = model.predict(input_scaled)
print("Prediction:", prediction)

if (prediction[0]==0):
  print("The person is not depressed")
else:
  print("The person is depressed")
```
